```python
def get_dir(cwd, group):
    '''Input current directory, sub folders, and return directory of all data files'''
    file_dirs = []
    for g in group:
        sub_dir = os.path.join(cwd, g)
        file_dir = [join(sub_dir,file) for file in listdir(sub_dir) if isfile(join(sub_dir, file))]
        file_dirs = file_dirs + file_dir
    return file_dirs

def read_data(file_dirs, group):
    '''Read directory and get all files as dfs'''
    dfs = []
    for file_dir in file_dirs:
        df = pd.read_csv(file_dir)
        date = re.sub('.csv(.*?)','',re.sub('(.*?)_','',file_dir))
        df['date'] = datetime.strptime(date, '%y-%m-%d')
        a = (g for g in group if g in file_dir)
        df['city'] = a
        dfs.append(df)
    return dfs
```

```python
dfs = read_data(get_dir(cwd,cities),cities)

#merge all
df = pd.concat(dfs, axis=0, ignore_index=True, sort=False)
df.shape
```
